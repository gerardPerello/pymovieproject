```python
# Load modules
import pandas as pd
import os
import requests
import json
import datetime as dt
import pandas as pd
from time import sleep
import os
from ..credentials import credentials
```

```python
# Get currency codes available in API 
file_path = '../data/currencies/currencies.json' # From website
df = pd.read_json(file_path)

currencies = ""
for a in df['code'].values:
    currencies += a + ","

print(currencies)
```

```python
import credentials

# API URL
url = 'http://api.currencylayer.com/historical'

# 500 days
start_date = dt.date(2011, 1, 1)
end_date = dt.date(2012, 5, 15)
delta = dt.timedelta(days=1)

# Loop through each day and get conversion rates to USD for every currency
data = []
while start_date <= end_date:
    print(start_date)
    
    params = {
        'access_key' : os.getenv('API_KEY'),
        'source' : 'USD',
        'currencies' : currencies,
        'date' : start_date
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        json_data = response.json()
        data.append(json_data)
    else:
        print('Error connecting to the API.')
        print(response)

    start_date += delta

    sleep(1)

print(pd.json_normalize(data))
```

# SEND CURRENCIES TO SNOWFLAKE

```python
curr_df = pd.read_json('../data/currencies/currencies.json')
curr_df.index += 1
curr_df.head()
```

```python
# Maps to create currency ids
code_to_id = {}
for ix, row in curr_df.iterrows():
    code_to_id[row['code']] = ix

# code_to_id
```

```python
import credentials
import snowflake.connector

def connect_snowflake():
    conn = snowflake.connector.connect(
        # user=os.getenv('SNOWSQL_USR'), 
        # password=os.getenv('SNOWSQL_PWD'), 
        # account=os.getenv('SNOWSQL_ACC'),
        # warehouse=os.getenv('SNOWSQL_WH'), 
        # database=os.getenv('SNOWSQL_DB'),
        # schema=os.getenv('SNOWSQL_SCH') 
    ) 
    print("Snowflake connection successfully.")
    return conn
```

```python
curr_df.columns
```

```python
# Send currency data

for ix, row in curr_df.iterrows():
    connection = connect_snowflake()
    cursor = connection.cursor()

    try:
        query = "INSERT INTO CURRENCIES (C_ID, C_CODE, C_NAME, C_COUNTRY, C_CONTINENT) VALUES (%s, %s, %s, %s, %s)" 
        
        cursor.execute(query, (ix, row['code'], row['name'], row['country'], row['continent']))
        connection.commit()

        print({'message': 'Currencies created successfully'}) 
    except Exception as e:
        print({'error': str(e)})
    finally:
        cursor.close()
        connection.close()
```

# Send forex history to SF

```python
forex_df = df_melted
forex_df['currency_id'] = forex_df['currency_code'].replace(code_to_id)
forex_df
```

```python
# Currency code doesn't go into this table
forex_df_new = forex_df.drop(columns=['currency_code'])
print(len(forex_df_new))

# Remove nulls
forex_df_cleaned = forex_df_new.where(pd.notna(forex_df_new), )
print(len(forex_df_cleaned))
forex_df_filtered = forex_df_cleaned.dropna(subset=['timestamp', 'date', 'currency_id', 'value_to_dollar'])
print(len(forex_df_filtered))
final_list = forex_df_filtered.to_dict(orient='records')
```

```python
# SEND TO SNOWFLAKE IN BATCHES 

batch_size = 10000  # You can adjust this number based on your specific use case

for i in range(0, len(final_list), batch_size):
    batch = final_list[i:i+batch_size]
    placeholders = ', '.join(['(%s, %s, %s, %s)'] * len(batch))
    query = f"INSERT INTO FOREX_HISTORY (FH_timestamp_id, FH_date, FH_currency_id, FH_value_to_dollar) VALUES {placeholders}"
    
    params = []
    for data in batch:
        params.extend([data['timestamp'], data['date'], data['currency_id'], data['value_to_dollar']])
    
    cursor.execute(query, params)
    connection.commit()

cursor.close()
connection.close()
```
